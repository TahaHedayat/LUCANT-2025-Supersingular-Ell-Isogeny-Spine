```python
import csv

center_file = 'center012925.csv'
# This data has the header: 'p', 'size of spine', 'size of center','size of graph', 'Fp vertices in center'

# Initialize lists to store the pairs
spine_pairs = []
center_pairs = []
graph_size_pairs = []
fp_center_pairs = []

# Read the CSV file
with open(center_file, mode='r', newline='') as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        p = row['p']
        size_of_spine = row['size of spine']
        size_of_center = row['size of center']
        graph_size = row['size of graph']
        fp_vertices_in_center = row['Fp vertices in center']
        
        
        # Store as tuples in respective lists
        spine_pairs.append((ZZ(p), ZZ(size_of_spine)))
        center_pairs.append((ZZ(p), ZZ(size_of_center)))
        graph_size_pairs.append((ZZ(p), ZZ(graph_size)))
        fp_center_pairs.append((ZZ(p), ZZ(fp_vertices_in_center)))

# Print results (optional)
#print("Pairs (p, size of spine):", spine_pairs)
#print("Pairs (p, Fp vertices in center):", fp_center_pairs)
```
